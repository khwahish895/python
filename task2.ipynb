```python
import os
import sys
import json
import csv
import shutil
import time
import datetime
import threading
import tkinter as tk
from tkinter import messagebox, filedialog
import requests
from bs4 import BeautifulSoup
import pandas as pd
import qrcode
import speedtest
from plyer import notification
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# =============================================================================
# 1. TRANSLATE TEXT AUTOMATICALLY
# =============================================================================

def translate_text():
    """Google Translate with error handling"""
    try:
        from googletrans import Translator
        translator = Translator()
        
        text = input("Enter text to translate: ")
        dest_lang = input("Destination language (e.g., 'en', 'fr', 'es'): ").lower()
        
        result = translator.translate(text, dest=dest_lang)
        print(f"Original: {text}")
        print(f"Translated: {result.text}")
        print(f"Source language: {result.src}")
        
    except Exception as e:
        print(f"Translation error: {e}")
        print("Make sure you have internet connection and correct language code")

# =============================================================================
# 2. BIRTHDAY REMINDER APP
# =============================================================================

class BirthdayReminder:
    def _init_(self):
        self.birthdays_file = "birthdays.json"
        self.load_birthdays()
    
    def load_birthdays(self):
        """Load birthdays from file"""
        try:
            with open(self.birthdays_file, 'r') as f:
                self.birthdays = json.load(f)
        except FileNotFoundError:
            self.birthdays = {}
    
    def save_birthdays(self):
        """Save birthdays to file"""
        with open(self.birthdays_file, 'w') as f:
            json.dump(self.birthdays, f, indent=2)
    
    def add_birthday(self, name, date):
        """Add a birthday (date format: YYYY-MM-DD)"""
        self.birthdays[name] = date
        self.save_birthdays()
        print(f"Birthday added for {name}: {date}")
    
    def check_birthdays(self):
        """Check for upcoming birthdays"""
        today = datetime.date.today()
        upcoming = []
        
        for name, birthday in self.birthdays.items():
            try:
                birth_date = datetime.datetime.strptime(birthday, "%Y-%m-%d").date()
                # Check if birthday is in next 7 days
                this_year = birth_date.replace(year=today.year)
                if this_year < today:
                    this_year = birth_date.replace(year=today.year + 1)
                
                days_until = (this_year - today).days
                if 0 <= days_until <= 7:
                    upcoming.append((name, this_year, days_until))
            except ValueError:
                print(f"Invalid date format for {name}: {birthday}")
        
        return upcoming
    
    def send_notification(self, name, days):
        """Send desktop notification"""
        try:
            if days == 0:
                message = f"ðŸŽ‰ Today is {name}'s birthday!"
            else:
                message = f"ðŸŽ‚ {name}'s birthday is in {days} day(s)!"
            
            notification.notify(
                title="Birthday Reminder",
                message=message,
                timeout=10
            )
        except Exception as e:
            print(f"Notification error: {e}")
    
    def run_reminder(self):
        """Run birthday reminder"""
        print("Birthday Reminder App")
        print("1. Add birthday")
        print("2. Check upcoming birthdays")
        print("3. Exit")
        
        while True:
            choice = input("\nEnter choice (1-3): ")
            
            if choice == '1':
                name = input("Enter name: ")
                date = input("Enter birthday (YYYY-MM-DD): ")
                self.add_birthday(name, date)
            
            elif choice == '2':
                upcoming = self.check_birthdays()
                if upcoming:
                    print("\nUpcoming birthdays:")
                    for name, date, days in upcoming:
                        print(f"- {name}: {date} ({days} days)")
                        self.send_notification(name, days)
                else:
                    print("No upcoming birthdays in the next 7 days")
            
            elif choice == '3':
                break
            
            else:
                print("Invalid choice!")

# =============================================================================
# 3. AUTO-ORGANIZE FILES IN FOLDERS
# =============================================================================

class FileOrganizer:
    def _init_(self, source_folder):
        self.source_folder = source_folder
        self.file_types = {
            'Documents': ['.pdf', '.doc', '.docx', '.txt', '.rtf'],
            'Images': ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.svg'],
            'Videos': ['.mp4', '.avi', '.mkv', '.mov', '.wmv', '.flv'],
            'Audio': ['.mp3', '.wav', '.flac', '.aac', '.wma'],
            'Archives': ['.zip', '.rar', '.7z', '.tar', '.gz'],
            'Spreadsheets': ['.xlsx', '.xls', '.csv'],
            'Presentations': ['.ppt', '.pptx']
        }
    
    def create_folders(self):
        """Create destination folders if they don't exist"""
        for folder_name in self.file_types.keys():
            folder_path = os.path.join(self.source_folder, folder_name)
            os.makedirs(folder_path, exist_ok=True)
    
    def organize_files(self):
        """Organize files by type"""
        if not os.path.exists(self.source_folder):
            print(f"Source folder doesn't exist: {self.source_folder}")
            return
        
        self.create_folders()
        moved_files = 0
        
        for filename in os.listdir(self.source_folder):
            if os.path.isfile(os.path.join(self.source_folder, filename)):
                file_ext = os.path.splitext(filename)[1].lower()
                
                for folder_name, extensions in self.file_types.items():
                    if file_ext in extensions:
                        source_path = os.path.join(self.source_folder, filename)
                        dest_folder = os.path.join(self.source_folder, folder_name)
                        dest_path = os.path.join(dest_folder, filename)
                        
                        try:
                            shutil.move(source_path, dest_path)
                            print(f"Moved: {filename} -> {folder_name}/")
                            moved_files += 1
                        except Exception as e:
                            print(f"Error moving {filename}: {e}")
                        break
        
        print(f"\nOrganization complete! Moved {moved_files} files.")

# =============================================================================
# 4. PRICE TRACKER
# =============================================================================

class PriceTracker:
    def _init_(self):
        self.prices_file = "price_history.json"
        self.load_price_history()
    
    def load_price_history(self):
        """Load price history from file"""
        try:
            with open(self.prices_file, 'r') as f:
                self.price_history = json.load(f)
        except FileNotFoundError:
            self.price_history = {}
    
    def save_price_history(self):
        """Save price history to file"""
        with open(self.prices_file, 'w') as f:
            json.dump(self.price_history, f, indent=2)
    
    def get_price(self, url):
        """Get price from Amazon product page"""
        try:
            headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
            }
            
            response = requests.get(url, headers=headers)
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # Amazon price selectors (may need updates)
            price_selectors = [
                '.a-price-whole',
                '.a-price .a-offscreen',
                '#priceblock_dealprice',
                '#priceblock_ourprice',
                '.a-price-range'
            ]
            
            for selector in price_selectors:
                price_element = soup.select_one(selector)
                if price_element:
                    price_text = price_element.get_text().strip()
                    # Extract numeric value
                    import re
                    price_match = re.search(r'[\d,]+\.?\d*', price_text)
                    if price_match:
                        return float(price_match.group().replace(',', ''))
            
            return None
        
        except Exception as e:
            print(f"Error getting price: {e}")
            return None
    
    def track_price(self, product_name, url, target_price):
        """Track price for a product"""
        current_price = self.get_price(url)
        
        if current_price is None:
            print(f"Could not get price for {product_name}")
            return
        
        timestamp = datetime.datetime.now().isoformat()
        
        if product_name not in self.price_history:
            self.price_history[product_name] = {
                'url': url,
                'target_price': target_price,
                'price_history': []
            }
        
        self.price_history[product_name]['price_history'].append({
            'price': current_price,
            'timestamp': timestamp
        })
        
        self.save_price_history()
        
        print(f"{product_name}: ${current_price}")
        
        if current_price <= target_price:
            try:
                notification.notify(
                    title="Price Alert!",
                    message=f"{product_name} is now ${current_price} (Target: ${target_price})",
                    timeout=15
                )
            except:
                print(f"PRICE ALERT: {product_name} is now ${current_price}!")

# =============================================================================
# 5. NEWS HEADLINES READER
# =============================================================================

class NewsReader:
    def _init_(self, api_key=None):
        self.api_key = api_key or "your_newsapi_key_here"
        self.base_url = "https://newsapi.org/v2/top-headlines"
    
    def get_headlines(self, country='us', category='general'):
        """Get news headlines"""
        try:
            params = {
                'country': country,
                'category': category,
                'apiKey': self.api_key
            }
            
            response = requests.get(self.base_url, params=params)
            data = response.json()
            
            if data['status'] == 'ok':
                return data['articles'][:10]  # Get top 10
            else:
                print(f"API Error: {data.get('message', 'Unknown error')}")
                return []
        
        except Exception as e:
            print(f"Error fetching news: {e}")
            return []
    
    def read_headlines(self):
        """Read headlines aloud (text-to-speech)"""
        headlines = self.get_headlines()
        
        if not headlines:
            print("No headlines available")
            return
        
        print("ðŸ“° TOP NEWS HEADLINES:\n")
        
        for i, article in enumerate(headlines, 1):
            title = article['title']
            print(f"{i}. {title}")
            
            # Text-to-speech (Windows)
            try:
                import pyttsx3
                engine = pyttsx3.init()
                engine.say(title)
                engine.runAndWait()
            except ImportError:
                print("Install pyttsx3 for text-to-speech: pip install pyttsx3")
            except:
                pass

# =============================================================================
# 6. EXPENSE TRACKER
# =============================================================================

class ExpenseTracker:
    def _init_(self):
        self.expenses_file = "expenses.csv"
        self.load_expenses()
    
    def load_expenses(self):
        """Load expenses from CSV file"""
        try:
            self.df = pd.read_csv(self.expenses_file)
        except FileNotFoundError:
            self.df = pd.DataFrame(columns=['Date', 'Category', 'Description', 'Amount'])
    
    def save_expenses(self):
        """Save expenses to CSV file"""
        self.df.to_csv(self.expenses_file, index=False)
    
    def add_expense(self, category, description, amount):
        """Add a new expense"""
        new_expense = {
            'Date': datetime.date.today().isoformat(),
            'Category': category,
            'Description': description,
            'Amount': float(amount)
        }
        
        self.df = pd.concat([self.df, pd.DataFrame([new_expense])], ignore_index=True)
        self.save_expenses()
        print(f"Expense added: {category} - ${amount}")
    
    def view_expenses(self):
        """View all expenses"""
        if self.df.empty:
            print("No expenses recorded yet.")
            return
        
        print("\nðŸ“Š EXPENSE SUMMARY:")
        print(self.df.to_string(index=False))
        
        print(f"\nTotal expenses: ${self.df['Amount'].sum():.2f}")
        
        # Category-wise summary
        category_summary = self.df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
        print("\nðŸ“ˆ Category-wise expenses:")
        for category, amount in category_summary.items():
            print(f"  {category}: ${amount:.2f}")
    
    def run_tracker(self):
        """Run expense tracker"""
        print("ðŸ’° Expense Tracker")
        print("1. Add expense")
        print("2. View expenses")
        print("3. Exit")
        
        while True:
            choice = input("\nEnter choice (1-3): ")
            
            if choice == '1':
                category = input("Category: ")
                description = input("Description: ")
                try:
                    amount = float(input("Amount: $"))
                    self.add_expense(category, description, amount)
                except ValueError:
                    print("Invalid amount!")
            
            elif choice == '2':
                self.view_expenses()
            
            elif choice == '3':
                break
            
            else:
                print("Invalid choice!")

# =============================================================================
# 7. QUIZ GENERATOR
# =============================================================================

class QuizGenerator:
    def _init_(self):
        self.api_url = "https://opentdb.com/api.php"
    
    def get_questions(self, amount=10, difficulty='medium'):
        """Get quiz questions from Open Trivia DB"""
        try:
            params = {
                'amount': amount,
                'difficulty': difficulty,
                'type': 'multiple'
            }
            
            response = requests.get(self.api_url, params=params)
            data = response.json()
            
            if data['response_code'] == 0:
                return data['results']
            else:
                print("Error fetching questions")
                return []
        
        except Exception as e:
            print(f"Error: {e}")
            return []
    
    def run_quiz(self):
        """Run interactive quiz"""
        print("ðŸ§  Random Quiz Generator")
        questions = self.get_questions()
        
        if not questions:
            print("Could not fetch questions. Check internet connection.")
            return
        
        score = 0
        
        for i, q in enumerate(questions, 1):
            print(f"\n--- Question {i}/{len(questions)} ---")
            print(f"Category: {q['category']}")
            print(f"Difficulty: {q['difficulty']}")
            print(f"\nQ: {q['question']}")
            
            # Prepare options
            options = q['incorrect_answers'] + [q['correct_answer']]
            import random
            random.shuffle(options)
            
            # Display options
            for j, option in enumerate(options, 1):
                print(f"{j}. {option}")
            
            # Get user answer
            try:
                user_choice = int(input("\nYour answer (1-4): ")) - 1
                if 0 <= user_choice < len(options):
                    if options[user_choice] == q['correct_answer']:
                        print("âœ… Correct!")
                        score += 1
                    else:
                        print(f"âŒ Wrong! Correct answer: {q['correct_answer']}")
                else:
                    print("Invalid choice!")
            except ValueError:
                print("Invalid input!")
        
        print(f"\nðŸŽ¯ Quiz Complete! Your score: {score}/{len(questions)} ({score/len(questions)*100:.1f}%)")

# =============================================================================
# 8. QR CODE GENERATOR
# =============================================================================

class QRCodeGenerator:
    def generate_qr(self, data, filename="qr_code.png"):
        """Generate QR code"""
        try:
            qr = qrcode.QRCode(
                version=1,
                error_correction=qrcode.constants.ERROR_CORRECT_L,
                box_size=10,
                border=4,
            )
            qr.add_data(data)
            qr.make(fit=True)
            
            img = qr.make_image(fill_color="black", back_color="white")
            img.save(filename)
            print(f"QR code saved as: {filename}")
            
        except Exception as e:
            print(f"Error generating QR code: {e}")
    
    def run_generator(self):
        """Run QR code generator"""
        print("ðŸ“± QR Code Generator")
        
        while True:
            print("\n1. Generate QR for text")
            print("2. Generate QR for URL")
            print("3. Exit")
            
            choice = input("Enter choice (1-3): ")
            
            if choice == '1':
                text = input("Enter text: ")
                filename = input("Filename (or press Enter for default): ") or "text_qr.png"
                self.generate_qr(text, filename)
            
            elif choice == '2':
                url = input("Enter URL: ")
                filename = input("Filename (or press Enter for default): ") or "url_qr.png"
                self.generate_qr(url, filename)
            
            elif choice == '3':
                break
            
            else:
                print("Invalid choice!")

# =============================================================================
# 9. CURRENCY CONVERTER
# =============================================================================

class CurrencyConverter:
    def _init_(self):
        self.api_url = "https://api.exchangerate-api.com/v4/latest/"
    
    def get_exchange_rate(self, base_currency, target_currency):
        """Get exchange rate"""
        try:
            response = requests.get(f"{self.api_url}{base_currency}")
            data = response.json()
            
            if target_currency in data['rates']:
                return data['rates'][target_currency]
            else:
                print(f"Currency {target_currency} not found")
                return None
        
        except Exception as e:
            print(f"Error fetching exchange rate: {e}")
            return None
    
    def convert_currency(self, amount, from_currency, to_currency):
        """Convert currency"""
        rate = self.get_exchange_rate(from_currency, to_currency)
        
        if rate:
            converted_amount = amount * rate
            print(f"{amount} {from_currency} = {converted_amount:.2f} {to_currency}")
            return converted_amount
        
        return None
    
    def run_converter(self):
        """Run currency converter"""
        print("ðŸ’± Currency Converter")
        
        while True:
            print("\n1. Convert currency")
            print("2. Exit")
            
            choice = input("Enter choice (1-2): ")
            
            if choice == '1':
                try:
                    amount = float(input("Enter amount: "))
                    from_curr = input("From currency (e.g., USD): ").upper()
                    to_curr = input("To currency (e.g., EUR): ").upper()
                    
                    self.convert_currency(amount, from_curr, to_curr)
                    
                except ValueError:
                    print("Invalid amount!")
            
            elif choice == '2':
                break
            
            else:
                print("Invalid choice!")

# =============================================================================
# 10. INTERNET SPEED CHECKER
# =============================================================================

class SpeedChecker:
    def check_speed(self):
        """Check internet speed"""
        try:
            print("ðŸŒ Testing internet speed...")
            st = speedtest.Speedtest()
            
            print("Finding best server...")
            st.get_best_server()
            
            print("Testing download speed...")
            download_speed = st.download() / 1_000_000  # Convert to Mbps
            
            print("Testing upload speed...")
            upload_speed = st.upload() / 1_000_000  # Convert to Mbps
            
            print("Getting ping...")
            ping = st.results.ping
            
            print(f"\nðŸ“Š Speed Test Results:")
            print(f"Download: {download_speed:.2f} Mbps")
            print(f"Upload: {upload_speed:.2f} Mbps")
            print(f"Ping: {ping:.2f} ms")
            
            # Save results
            timestamp = datetime.datetime.now().isoformat()
            result = {
                'timestamp': timestamp,
                'download': download_speed,
                'upload': upload_speed,
                'ping': ping
            }
            
            try:
                with open('speed_history.json', 'r') as f:
                    history = json.load(f)
            except FileNotFoundError:
                history = []
            
            history.append(result)
            
            with open('speed_history.json', 'w') as f:
                json.dump(history, f, indent=2)
            
        except Exception as e:
            print(f"Error checking speed: {e}")

# =============================================================================
# 11. DIGITAL CLOCK GUI
# =============================================================================

class DigitalClock:
    def _init_(self):
        self.root = tk.Tk()
        self.root.title("Digital Clock")
        self.root.geometry("400x150")
        self.root.configure(bg='black')
        
        self.time_label = tk.Label(
            self.root,
            font=('Arial', 24, 'bold'),
            bg='black',
            fg='cyan'
        )
        self.time_label.pack(expand=True)
        
        self.date_label = tk.Label(
            self.root,
            font=('Arial', 12),
            bg='black',
            fg='white'
        )
        self.date_label.pack()
        
        self.update_time()
    
    def update_time(self):
        """Update time display"""
        now = datetime.datetime.now()
        time_string = now.strftime("%H:%M:%S")
        date_string = now.strftime("%A, %B %d, %Y")
        
        self.time_label.config(text=time_string)
        self.date_label.config(text=date_string)
        
        self.root.after(1000, self.update_time)
    
    def run(self):
        """Run the clock"""
        self.root.mainloop()

# =============================================================================
# 12. SHOPPING LIST APP
# =============================================================================

class ShoppingListApp:
    def _init_(self):
        self.root = tk.Tk()
        self.root.title("Shopping List")
        self.root.geometry("400x500")
        
        self.items = []
        self.setup_ui()
        self.load_list()
    
    def setup_ui(self):
        """Setup user interface"""
        # Title
        title_label = tk.Label(self.root, text="ðŸ› Shopping List", font=('Arial', 16, 'bold'))
        title_label.pack(pady=10)
        
        # Input frame
        input_frame = tk.Frame(self.root)
        input_frame.pack(pady=10)
        
        self.item_entry = tk.Entry(input_frame, width=30)
        self.item_entry.pack(side=tk.LEFT, padx=5)
        
        add_button = tk.Button(input_frame, text="Add Item", command=self.add_item)
        add_button.pack(side=tk.LEFT, padx=5)
        
        # List frame
        list_frame = tk.Frame(self.root)
        list_frame.pack(pady=10, fill=tk.BOTH, expand=True)
        
        # Scrollbar
        scrollbar = tk.Scrollbar(list_frame)
        scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
        
        # Listbox
        self.listbox = tk.Listbox(list_frame, yscrollcommand=scrollbar.set)
        self.listbox.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
        scrollbar.config(command=self.listbox.yview)
        
        # Buttons frame
        buttons_frame = tk.Frame(self.root)
        buttons_frame.pack(pady=10)
        
        delete_button = tk.Button(buttons_frame, text="Delete Selected", command=self.delete_item)
        delete_button.pack(side=tk.LEFT, padx=5)
        
        clear_button = tk.Button(buttons_frame, text="Clear All", command=self.clear_all)
        clear_button.pack(side=tk.LEFT, padx=5)
        
        save_button = tk.Button(buttons_frame, text="Save List", command=self.save_list)
        save_button.pack(side=tk.LEFT, padx=5)
        
        # Bind Enter key
        self.item_entry.bind('<Return>', lambda e: self.add_item())
    
    def add_item(self):
        """Add item to list"""
        item = self.item_entry.get().strip()
        if item:
            self.items.append(item)
            self.listbox.insert(tk.END, item)
            self.item_entry.delete(0, tk.END)
    
    def delete_item(self):
        """Delete selected item"""
        selection = self.listbox.curselection()
        if selection:
            index = selection[0]
            self.listbox.delete(index)
            del self.items[index]
    
    def clear_all(self):
        """Clear all items"""
        self.listbox.delete(0, tk.END)
        self.items.clear()
    
    def save_list(self):
        """Save list to file"""
        try:
            with open('shopping_list.txt', 'w') as f:
                for item in self.items:
                    f.write(f"{item}\n")
            messagebox.showinfo("Success", "Shopping list saved!")
        except Exception as e:
            messagebox.showerror("Error", f"Error saving list: {e}")
    
    def load_list(self):
        """Load list from file"""
        try:
            with open('shopping_list.txt', 'r') as f:
                self.items = [line.strip() for line in f.readlines()]
                for item in self.items:
                    self.listbox.insert(tk.END, item)
        except FileNotFoundError:
            pass
    
    def run(self):
        """Run the app"""
        self.root.mainloop()

# =============================================================================
# MAIN MENU - RUN ALL PROJECTS
# =============================================================================

def main_menu():
    """Main menu to run all projects"""
    print("ðŸš€ 25 Python Automation Projects")
    print("=" * 50)
    
    projects = {
        '1': ("Translate Text", translate_text),
        '2': ("Birthday Reminder", lambda: BirthdayReminder().run_reminder()),
        '3': ("File Organizer", lambda: FileOrganizer(input("Enter folder path: ")).organize_files()),
        '4': ("Price Tracker", lambda: PriceTracker().track_price(
            input("Product name: "),
            input("Product URL: "),
            float(input("Target price: $"))
        )),
        '5': ("News Reader", lambda: NewsReader().read_headlines()),
        '6': ("Expense Tracker", lambda: ExpenseTracker().run_tracker()),
        '7': ("Quiz Generator", lambda: QuizGenerator().run_quiz()),
        '8': ("QR Code Generator", lambda: QRCodeGenerator().run_generator()),
        '9': ("Currency Converter", lambda: CurrencyConverter().run_converter()),
        '10': ("Speed Checker", lambda: SpeedChecker().check_speed()),
        '11': ("Digital Clock", lambda: DigitalClock().run()),
        '12': ("Shopping List", lambda: ShoppingListApp().run()),
    }
    
    while True:
        print("\nðŸ“‹ Available Projects:")
        for key, (name, _) in projects.items():
            print(f"{key}. {name}")
        
        print("0. Exit")
        
        choice = input("\nEnter your choice: ")
        
        if choice == '0':
            print("Goodbye! ðŸ‘‹")
            break
        
        elif choice in projects:
            name, func = projects[choice]
            print(f"\nðŸŽ¯ Running: {name}")
            print("-" * 30)
            
            try:
                func()
            except Exception as e:
                print(f"Error running {name}: {e}")
            
            input("\nPress Enter to continue...")
        
        else:
            print("Invalid choice!")

if _name_ == "_main_":
    main_menu()

# =============================================================================
# ADDITIONAL UTILITY FUNCTIONS
# =============================================================================

def install_requirements():
    """Install required packages"""
    packages = [
        'googletrans==4.0.0-rc1',
        'requests',
        'beautifulsoup4',
        'newsapi-python',
        'plyer',
        'pandas',
        'openpyxl',
        'qrcode[pil]',
        'forex-python',
        'schedule',
        'pywhatkit',
        'speedtest-cli',
        'imdbpy',
        'pyttsx3'
    ]
    
    print("Installing required packages...")
    for package in packages:
        try:
            os.system(f"
```
