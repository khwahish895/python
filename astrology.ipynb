```python
import os
import gradio as gr
import google.generativeai as genai
from datetime import datetime

# Configure Gemini API key
genai.configure(api_key="AIzaSyDrx7j5XPIztMz274t-ItjxSE55sr6Q39g")
model = genai.GenerativeModel('gemini-2.5-flash')

# Astrology Prediction Function
def get_astrology_prediction(name, dob, zodiac, question):
    try:
        # Validate date
        dob_obj = datetime.strptime(dob, "%Y-%m-%d")
        prompt = f"""
        You are an expert astrologer.
        User name: {name}
        Date of birth: {dob_obj.strftime('%B %d, %Y')}
        Zodiac sign: {zodiac}
        Question: {question}

        Based on this information, provide a detailed astrological prediction and guidance.
        """

        response = model.generate_content(prompt)
        return response.text
    except ValueError:
        return "❌ Please enter the date in YYYY-MM-DD format."
    except Exception as e:
        return f"❌ Error: {str(e)}"

# Gradio UI
with gr.Blocks(css="""
body { background-color: #f8f4ff; font-family: 'Arial', sans-serif; }
textarea, input, select, button { 
    background-color: #fff0f5; 
    border: 1px solid #d8bfd8; 
    font-size: 16px; 
    color: #4b0082; 
}
button:hover {
    background-color: #dda0dd;
}
""") as demo:

    gr.Markdown("<h1 style='color:#800080;'>🔮 Astrology Predictor</h1>")
    gr.Markdown("Enter your details below to receive an astrological prediction.")

    with gr.Row():
        name = gr.Textbox(label="Your Name")
        dob = gr.Textbox(label="Date of Birth (YYYY-MM-DD)")
    
    zodiac = gr.Dropdown(
        ["Aries", "Taurus", "Gemini", "Cancer", "Leo", "Virgo",
         "Libra", "Scorpio", "Sagittarius", "Capricorn", "Aquarius", "Pisces"],
        label="Zodiac Sign"
    )
    question = gr.Textbox(label="What's on your mind today?", lines=3, placeholder="Type your question here...")
    predict_btn = gr.Button("✨ Get Prediction")
    output = gr.Textbox(label="Your Astrology Prediction", lines=10)

    predict_btn.click(fn=get_astrology_prediction, 
                      inputs=[name, dob, zodiac, question], 
                      outputs=output)

# Launch with share=True
demo.launch(share=True)
```
